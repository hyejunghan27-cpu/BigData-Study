```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['합격핑', '만점핑', '열심핑', '도전핑', '재수핑'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85, 90, 78, np.nan, 95],
})
df
```

```python
cond = df['Age'] > 25
cond
```

```python
print("sum:", sum(cond)) # F->0 , T->1
print("len:", len(cond))
```

```python
print(len(df[cond]))
print(len(df[df['Age']>25]))
```

```python
df[cond].count()
```

```python
cond2 = df['Score'] > 80
print("sum:", sum(cond2))
print("len:", len(cond2))
```

```python
print("len:", len(df[cond2]))
```
